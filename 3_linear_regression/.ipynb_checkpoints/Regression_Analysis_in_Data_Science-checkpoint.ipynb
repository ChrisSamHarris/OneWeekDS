# CODING TASK #4: TRAIN & EVALUATE AN XG-BOOST ALGORITHM

```python
# Train an XGBoost regressor model 

import xgboost as xgb

# learning_rate = How aggressively you would like to train the model 
# max_depth = depth of the tree for the model 
model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)

model.fit(x_train, y_train)
```

```python
# predict the score of the trained model using the testing dataset

result = model.score(x_test, y_test)
print("Accuracy : {}".format(result))
```

```python
# make predictions on the test data
y_predict = model.predict(x_test)
y_predict[:5]
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2) 
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
life_exp_df = pd.read_csv("Life_Expectancy_Data.csv")
life_exp_df.head()
```

```python
life_exp_df.shape
```

```python
missing_values = life_exp_df.isnull()
missing_values.shape
life_exp_df.isnull().sum()
```

```python
life_exp_rm_missing = life_exp_df.dropna()
life_exp_rm_missing.isnull().sum()
```

```python
life_exp_rm_missing.info()
```

```python
life_exp_rm_missing.keys()
```

```python
min_le = life_exp_rm_missing['Life expectancy '].min().round(1)
max_le = life_exp_rm_missing['Life expectancy '].max().round(1)
mean_le = life_exp_rm_missing['Life expectancy '].mean().round(1)
print(f'min le: {min_le} | max le: {max_le} | mean le: {mean_le}')
```

```python
result = life_exp_rm_missing[life_exp_rm_missing.isin(['Developing'])]
result
```

```python
life_exp_rm_missing['Status'].unique()
```

```python
life_exp_rm_missing['Status'].nunique()
```

```python
sns.scatterplot(x='Income composition of resources', y = 'Life expectancy ', hue = 'Status', data = life_exp_rm_missing)
```

```python
# only 2 values in Status so one-hot encoding them to make them numeric 
life_exp_refactored = pd.get_dummies(life_exp_rm_missing, columns = ['Status'])
life_exp_refactored.head()
```

```python
plt.figure(figsize = (20,10))
corr_matrix = life_exp_refactored.corr()
sns.heatmap(corr_matrix, annot=True)
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb

no_le = life_exp_refactored.drop(columns = ['Life expectancy '])
le_results = life_exp_refactored[['Life expectancy ']]

x = np.array(no_le).astype('float32')
y = np.array(le_results).astype('float32')

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)

model.fit(x_train, y_train)
```

```python
model_accuracy = model.score(x_test, y_test)
model_accuracy
```

```python
y_predict = model.predict(x_test)
y_predict[:5]
```

```python
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = x_test.shape[1]
n = len(x_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```

```python
# Checking the unique values in country to consider it as a categorical variable
life_expectancy_df['Status'].nunique()
```

```python
# Perform one-hot encoding
life_expectancy_df = pd.get_dummies(life_expectancy_df, columns = ['Status'])
```

```python
life_expectancy_df
```

```python
# Check the number of null values for the columns having null values
life_expectancy_df.isnull().sum()[np.where(life_expectancy_df.isnull().sum() != 0)[0]]
```

```python
# Since most of the are continous values we fill them with mean
life_expectancy_df = life_expectancy_df.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
life_expectancy_df.isnull().sum()[np.where(life_expectancy_df.isnull().sum() != 0)[0]]
```

```python
# Create train and test data

X = life_expectancy_df.drop(columns = ['Life expectancy '])
y = life_expectancy_df[['Life expectancy ']]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
# Convert the data type to float32

X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
# Only take the numerical variables and scale them
X 
```

```python
# split the data into test and train sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
# Train an XGBoost regressor model 

import xgboost as xgb
model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)
model.fit(X_train, y_train)
```

```python
model_accuracy = model.score(X_test, y_test)
model_accuracy
```

```python
# Make prediction
y_predict = model.predict(X_test)
y_predict
```

```python
# Plot the scaled result

plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.show()
```

```python
# Plot the KPIs

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2) 
```
